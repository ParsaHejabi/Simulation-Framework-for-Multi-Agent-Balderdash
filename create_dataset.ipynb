```python
def clean_definition_text(text):
    """
    Cleans the definition text by removing or replacing special tokens.
    """
    # Remove bold, italics, smallcaps, and other formatting
    text = re.sub(r"\{/?it\}", "", text)  # Remove italic tags
    text = re.sub(r"\{/?b\}", "", text)  # Remove bold tags
    text = re.sub(r"\{/?sc\}", "", text)  # Remove smallcaps tags
    text = re.sub(r"\{/?sup\}", "", text)  # Remove superscript tags
    text = re.sub(r"\{/?inf\}", "", text)  # Remove subscript tags

    # Replace punctuation tokens
    text = re.sub(r"\{bc\}", "", text)  # Replace {bc} with colon and space
    text = re.sub(r"\{ldquo\}", '"', text)  # Replace left double quote
    text = re.sub(r"\{rdquo\}", '"', text)  # Replace right double quote

    # Replace link and cross-reference tokens, showing only the linked term
    text = re.sub(r"\{[a-z_]+_link\|([^|}]+)\|?[^}]*\}", r"\1", text)

    # Replace cross-reference tokens like {dxt|...|...|}, removing extra details
    text = re.sub(r"\{dxt\|([^|}]+)\|[^}]*\}", r"\1", text)

    # Replace synonymous cross-reference {sx|term|...} with 'term'
    text = re.sub(r"\{sx\|([^|}]+)\|[^}]*\}", r"\1", text)

    # Handle nested complex tokens like {gloss|...} and similar
    text = re.sub(r"\{[^}]+\}([^{}]+)\{\/[^}]+\}", r"\1", text)

    return text.strip()
```

```python
df["def"] = df["def"].progress_apply(clean_definition_text)

df
```

```python
# Create a pandas dataframe with columns "words" "def" and "POS" and fill the "words" column with the words
df = pd.DataFrame(words, columns=["words"])
df["def"] = ""
df["POS"] = ""
df
```

```python
def get_word_definition_and_pos(word, api_key):
    """Retrieve detailed definitions and their corresponding parts of speech for a given word from Merriam-Webster API."""
    url = f"https://dictionaryapi.com/api/v3/references/collegiate/json/{word}?key={api_key}"
    response = requests.get(url)
    if response.status_code != 200:
        return None, None

    data = response.json()
    definition_pos_pairs = []

    for entry in data:
        if "meta" in entry and "id" in entry["meta"] and word in entry["meta"]["id"]:
            pos = entry.get("fl", None)
            if "def" in entry:
                # Extract definitions from each sense in 'sseq'
                for sense_block in entry["def"]:
                    for sense in sense_block.get("sseq", []):
                        for sense_detail in sense:
                            if sense_detail[0] == "sense":
                                definition_text = sense_detail[1].get("dt", [])
                                # Collect each textual definition
                                for text in definition_text:
                                    if text[0] == "text":
                                        clean_definition = text[1].strip()
                                        definition_pos_pairs.append((clean_definition, pos))

    return definition_pos_pairs
```

```python
# Assume df is your DataFrame and it has a column named 'words'
progress_bar = tqdm(df.iterrows(), total=df.shape[0])

for index, row in progress_bar:
    word = row["words"]
    api_key = os.getenv("DICTIONARY_API_KEY")
    definition_pos_pairs = get_word_definition_and_pos(word, api_key)
    if definition_pos_pairs:
        # Separate definitions and POSs
        definitions = [pair[0] for pair in definition_pos_pairs]
        poss = [pair[1] for pair in definition_pos_pairs if pair[1] is not None]
        # make string version of definitions and POSs
        df.at[index, "def"] = definitions.__str__()
        df.at[index, "POS"] = poss.__str__()
    else:
        # Update progress bar to show the word was not found
        progress_bar.set_postfix({"word": word, "status": "not found"})
        df.at[index, "def"] = -1  # Use None or a specific string if -1 is not appropriate
        df.at[index, "POS"] = -1  # Same as above

df
```

```python
def get_word_definition_and_pos(word, api_key):
    """
    Makes a request to the Merriam-Webster Dictionary API for a given word and returns the first definition
    and part of speech (POS) if available.

    Parameters:
    word (str): The word for which the definition and POS are requested.
    api_key (str): Your Merriam-Webster API key.

    Returns:
    tuple: A tuple containing the first definition and POS of the word. If multiple
    definitions or POS tags are present, only the first is returned. If no definition
    or POS is found, None is returned for each.
    """
    url = f"https://dictionaryapi.com/api/v3/references/collegiate/json/{word}?key={api_key}"
    response = requests.get(url)
    response_data = response.json()

    # Check if the response contains data and extract the first definition and POS
    if (
        response_data
        and isinstance(response_data, list)
        and "shortdef" in response_data[0]
        and "fl" in response_data[0]
    ):
        first_definition = response_data[0]["shortdef"][0] if response_data[0]["shortdef"] else None
        pos = response_data[0]["fl"] if response_data[0]["fl"] else None
        return first_definition, pos
    else:
        # Handle cases where the word might not be found or API response is unexpected
        return None, None
```

```python
# Make a single dataframe containing all the words from both CSV files
balderdash_words_df = pd.concat([balderdash_words1_df, balderdash_words2_df], ignore_index=True)
balderdash_words_df
```

```python
# Drop the POS column
balderdash_words_df = balderdash_words_df.drop(columns=["POS"])
balderdash_words_df
```

```python
# Rename "words" to "word" and "def" to "dictionary_definition"
balderdash_words_df = balderdash_words_df.rename(columns={"words": "word", "def": "dictionary_definition"})
# Clear the "dictionary_definition" column
balderdash_words_df["dictionary_definition"] = None
# Create a new column "pos" and fill it with None
balderdash_words_df["pos"] = None
balderdash_words_df
```

```python
# For each word in the dataframe, get the definition and POS from the Merriam-Webster API
# and update the corresponding columns in the dataframe
progress_bar = tqdm(balderdash_words_df.iterrows(), total=balderdash_words_df.shape[0])

for index, row in progress_bar:
    word = row["word"]
    definition, pos = get_word_definition_and_pos(word, os.getenv("DICTIONARY_API_KEY"))
    if definition is not None:
        balderdash_words_df.at[index, "dictionary_definition"] = definition
        balderdash_words_df.at[index, "pos"] = pos
    else:
        # Write in tqdm progress bar: "Word {word} not found in the dictionary."
        progress_bar.set_postfix({"word": word, "status": "not found"})
        # Write -1 in the dictionary_definition and pos column if the word is not found
        balderdash_words_df.at[index, "dictionary_definition"] = -1
        balderdash_words_df.at[index, "pos"] = -1

balderdash_words_df
```

```python
# Save the updated dataframe to a new CSV file
balderdash_words_with_definitions_path = os.path.join(
    os.getcwd(), "data", "balderdash_words_with_definitions.csv"
)
balderdash_words_df.to_csv(balderdash_words_with_definitions_path, index=False)
```
